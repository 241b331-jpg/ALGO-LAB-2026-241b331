# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    predecessors = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')
        predecessors[node] = None

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None:
            break

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node

    return distances, predecessors
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
# Dijkstra Algorithm Implementation in Python

def dijkstra(graph, start):
    # Initialize distances and predecessors
    distances = {node: float('inf') for node in graph}
    predecessors = {node: None for node in graph}
    visited = set()

    # Distance to the source node is 0
    distances[start] = 0

    while len(visited) < len(graph):
        # Select the unvisited node with the smallest distance
        current_node = None
        current_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < current_distance:
                current_distance = distances[node]
                current_node = node

        # If no node is reachable, break
        if current_node is None:
            break

        # Mark the node as visited
        visited.add(current_node)

        # Update distances of neighboring nodes
        for neighbor, weight in graph[current_node].items():
            new_distance = distances[current_node] + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = current_node

    return distances, predecessors


# Function to reconstruct the shortest path
def get_path(predecessors, start, end):
    path = []
    current = end

    while current is not None:
        path.insert(0, current)
        current = predecessors[current]

    if path[0] == start:
        return " → ".join(path)
    else:
        return "No path exists"


# Function to take user input for the graph
def input_graph():
    graph = {}
    num_nodes = int(input("Enter number of nodes: "))

    print("\nEnter node names:")
    nodes = [input(f"Node {i + 1}: ") for i in range(num_nodes)]

    # Initialize dictionary
    for node in nodes:
        graph[node] = {}

    print("\nEnter edges (enter 'done' to stop):")
    while True:
        u = input("From: ")
        if u.lower() == "done":
            break
        v = input("To: ")
        w = int(input("Weight: "))

        # Add edge (undirected graph)
        graph[u][v] = w
        graph[v][u] = w

    return graph



print("===== DIJKSTRA SHORTEST PATH ALGORITHM =====")

# User input graph
graph = input_graph()

# Adding two more nodes manually (optional demonstration)
print("\nAdding two additional nodes: X and Y")
graph['X'] = {'Y': 7}
graph['Y'] = {'X': 7}

# Connect them to existing nodes if desired
if len(graph) > 2:
    first_node = list(graph.keys())[0]
    graph[first_node]['X'] = 6
    graph['X'][first_node] = 6

# Display graph
print("\nGraph Representation:")
for node in graph:
    print(f"{node}: {graph[node]}")

# Test with different source nodes
while True:
    start = input("\nEnter source node: ")
    if start not in graph:
        print("Invalid source node! Try again.")
        continue

    distances, predecessors = dijkstra(graph, start)

    print(f"\nShortest distances from source '{start}':")
    for node in graph:
        print(f"Distance to {node}: {distances[node]}")

    print("\nShortest Paths:")
    for node in graph:
        path = get_path(predecessors, start, node)
        print(f"{start} → {node}: {path} (Cost: {distances[node]})")

    choice = input("\nDo you want to try another source node? (yes/no): ")
    if choice.lower() != 'yes':
        break
```
