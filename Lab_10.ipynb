# Warshall algorithm to find reachability between all pairs of vertices in a graph.
Problem Description:
Understand transitive closure of a graph. Use Python to compute reachability using matrix method.

```python
def warshall(graph):

    n = len(graph)

    for k in range(n):
        for i in range(n):
            for j in range(n):
                graph[i][j] = graph[i][j] or (graph[i][k] and graph[k][j])

    return graph
```

## Find shortest paths between all pairs of vertices using Floyd Warshall algorithm.

#Understand all pairs shortest path problem. Implement dynamic programming approach in Python.

* Given a weighted graph with positive or negative edge weights, find
shortest distances between every pair of vertices.

# Theory
Floyd Warshall algorithm computes shortest paths between all pairs of nodes. It updates the distance matrix by checking if a shorter path exists through an intermediate vertex. Works with negative weights but fails with negative cycles.

Algorithm
1. Represent graph using distance matrix.
2. Initialize matrix with given weights.
3. Set distance of node to itself as 0.
4. For each intermediate vertex k
5. Update distance[i][j] = min(distance[i][j], distance[i][k] + distance[k][j])
6. Repeat for all vertices

```python
def floyd_warshall(graph):

    n = len(graph)

    dist = []
    for i in range(n):
        dist.append(graph[i][:])

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist

```

```python
INF = float('inf')

graph = [
[0, 3, INF, 7],
[8, 0, 2, INF],
[5, INF, 0, 1],
[2, INF, INF, 0]
]

result = floyd_warshall(graph)

print("Shortest distance matrix:")

for row in result:
    print(row)
```

```python


INF = float('inf')

def input_graph(n):
    print(f"Enter adjacency matrix ({n}x{n}):")
    graph = []
    for i in range(n):
        row = list(map(int, input(f"Row {i+1}: ").split()))
        graph.append(row)
    return graph


def warshall(graph):
    n = len(graph)
    reach = [row[:] for row in graph]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                reach[i][j] = reach[i][j] or (reach[i][k] and reach[k][j])

    return reach

def floyd(graph):
    n = len(graph)
    dist = [row[:] for row in graph]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])

    return dist


def display_paths(reach):
    n = len(reach)
    print("\nPath Existence:")
    for i in range(n):
        for j in range(n):
            if reach[i][j]:
                print(f"Path exists from {i} → {j}")
            else:
                print(f"No path from {i} → {j}")


n = 5
graph = input_graph(n)


transitive_closure = warshall(graph)

print("\nTransitive Closure Matrix (Warshall):")
for row in transitive_closure:
    print(row)

display_paths(transitive_closure)


weighted_graph = []
for i in range(n):
    row = []
    for j in range(n):
        if i == j:
            row.append(0)
        elif graph[i][j] == 0:
            row.append(INF)
        else:
            row.append(graph[i][j])
    weighted_graph.append(row)


shortest_paths = floyd(weighted_graph)

print("\nShortest Path Matrix (Floyd):")
for row in shortest_paths:
    print(row)


print("\nComparison:")
print("Warshall → checks if path exists (0 or 1)")
print("Floyd → gives shortest path distances")
```

```python
INF = float('inf')


def input_graph(n):
    print(f"Enter weighted matrix ({n}x{n}) (use large number like 99999 for INF):")
    graph = []
    for i in range(n):
        row = list(map(int, input(f"Row {i+1}: ").split()))
        graph.append([INF if x == 99999 else x for x in row])
    return graph

def floyd_warshall(graph):
    n = len(graph)
    dist = [row[:] for row in graph]
    next_node = [[j if graph[i][j] != INF else None for j in range(n)] for i in range(n)]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] != INF and dist[k][j] != INF:
                    if dist[i][k] + dist[k][j] < dist[i][j]:
                        dist[i][j] = dist[i][k] + dist[k][j]
                        next_node[i][j] = next_node[i][k]

    return dist, next_node

def has_negative_cycle(dist):
    n = len(dist)
    for i in range(n):
        if dist[i][i] < 0:
            return True
    return False

def get_path(u, v, next_node):
    if next_node[u][v] is None:
        return None
    path = [u]
    while u != v:
        u = next_node[u][v]
        path.append(u)
    return path

def dijkstra(graph, src):
    n = len(graph)
    dist = [INF] * n
    visited = [False] * n
    dist[src] = 0

    for _ in range(n):
        u = min((d, i) for i, d in enumerate(dist) if not visited[i])[1]
        visited[u] = True

        for v in range(n):
            if graph[u][v] != INF and not visited[v]:
                if dist[u] + graph[u][v] < dist[v]:
                    dist[v] = dist[u] + graph[u][v]

    return dist


n = int(input("Enter number of vertices: "))
graph = input_graph(n)

dist, next_node = floyd_warshall(graph)

print("\nShortest Distance Matrix:")
for row in dist:
    print(row)

if has_negative_cycle(dist):
    print("\n⚠️ Negative cycle detected!")
else:
    print("\nNo negative cycle detected.")

u = int(input("\nEnter source vertex: "))
v = int(input("Enter destination vertex: "))

path = get_path(u, v, next_node)

if path:
    print(f"Shortest path from {u} to {v}: {' → '.join(map(str, path))}")
    print(f"Distance: {dist[u][v]}")
else:
    print("No path exists.")

src = int(input("\nEnter source for Dijkstra: "))
dijkstra_result = dijkstra(graph, src)

print("\nDijkstra distances from source:")
print(dijkstra_result)


print("\nComparison:")
print("Floyd-Warshall → All pairs shortest paths (handles negative weights)")
print("Dijkstra → Single source shortest path (fails with negative weights)")
```
